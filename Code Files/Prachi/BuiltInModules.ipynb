<h1>3. Date and Time</h1>

```python
from datetime import date, time, datetime

# Create a date object
today = date.today()
print("Today's date:", today)

# Create a specific date
specific_date = date(2023, 11, 6)
print("Specific date:", specific_date)

# Create a time object
specific_time = time(14, 30, 15)
print("Specific time:", specific_time)

# Create a datetime object
specific_datetime = datetime(2023, 11, 6, 14, 30, 15)
print("Specific datetime:", specific_datetime)
```

```python
now = datetime.now()
print("Current date and time:", now)

utc_now = datetime.utcnow()
print("Current UTC date and time:", utc_now)
```

```python
# Format a date or datetime as a string
formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
print("Formatted date:", formatted_date)

# Parse a string into a datetime object
parsed_date = datetime.strptime("2023-11-06 14:30:15", "%Y-%m-%d %H:%M:%S")
print("Parsed date:", parsed_date)
```

```python
#Date Arithmetic with timedelta
from datetime import timedelta

# Add 10 days to the current date
future_date = today + timedelta(days=10)
print("Future date:", future_date)

# Calculate time difference
time_difference = future_date - today
print("Days difference:", time_difference.days)
```

<h3>time Module</h3>
The time module is a lower-level module that primarily deals with time-related functions, such as sleeping, timestamps, and performance measurement.

```python
#Getting the Current Time as a Timestamp
import time

current_timestamp = time.time()
print("Current timestamp:", current_timestamp)
```

```python
#Converting Timestamps to Readable Dates
# Convert timestamp to a struct_time object in UTC
utc_time = time.gmtime(current_timestamp)
print("UTC time:", time.strftime("%Y-%m-%d %H:%M:%S", utc_time))

# Convert timestamp to a struct_time object in local time
local_time = time.localtime(current_timestamp)
print("Local time:", time.strftime("%Y-%m-%d %H:%M:%S", local_time))
```

```python
#Pausing Execution with sleep
print("Start sleep")
time.sleep(2)  # Pauses execution for 2 seconds
print("End sleep")
```

<h3>calendar Module</h3>
The calendar module provides functions to work with yearly and monthly calendars, checking for leap years, and working with weekdays.

```python
#Displaying a Calendar
import calendar

# Print a text calendar for a given month and year
print(calendar.month(2023, 11))

# Print a text calendar for a full year
print(calendar.calendar(2023))
```

```python
#Checking for Leap Year
is_leap = calendar.isleap(2024)
print("Is 2024 a leap year?", is_leap)
```

```python
#Getting the Day of the Week
# 0 = Monday, 6 = Sunday
day_of_week = calendar.weekday(2024, 11, 6)
print("Day of the week:", day_of_week)
```

```python
#Iterating Over Calendar Days
#The monthcalendar function returns a list of weeks for a given month.
# List of weeks in November 2023 (0 represents no day in that week slot)
november_calendar = calendar.monthcalendar(2023, 11)
print("November calendar:", november_calendar)
```

<h1>5. Network and Internet Protocols</h1>

```python
import http.client

# Use only the hostname in HTTPSConnection
conn = http.client.HTTPSConnection("github.com")

# Specify the path separately in the request
conn.request("GET", "/prachikabra121?tab=repositories")

# Get the response
response = conn.getresponse()
print(response.status, response.reason)
data = response.read()
print(data.decode("utf-8"))
```

```python
#urllib: Handles URL manipulation, downloading files, and working with HTTP.

from urllib import request
response = request.urlopen("https://github.com/prachikabra121?tab=repositories")

print(response.read())

```

```python
from urllib import request

# Enclose the URL in double quotes
url = "https://github.com/prachikabra121?tab=repositories"
response = request.urlopen(url)

# Read and print the response
data = response.read()
print(data.decode("utf-8"))
```
