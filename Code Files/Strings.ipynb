## String indexing


```python
b="abc12334"
print(b[0]) # Start
print(b[-1]) # End
print(b[:-2]) # before mentioned index
print(b[-2:]) # equal & after index
print(b[3:6])
print(b[3:-2])
print(b[3:12])
print(b.count("3"))
print(b.index("3"))
```
